```python
# Loading pre-trained embeddings
!pip install gensim
import gensim.downloader as api

# Handling numerical matrices
import numpy as np

# Visualization
import matplotlib.pyplot as plt

# Dimensionality reduction (t-SNE)
from sklearn.manifold import TSNE
```

```python
model = api.load("glove-wiki-gigaword-100")  # 100-dimensional embeddings

print("Vocabulary size:", len(model))

# Display example vector
print("Vector for 'computer':\n", model['computer'])
print("Vector shape:", model['computer'].shape)
```

```python
words = [
    # Animals
    "cat", "dog", "lion", "tiger", "elephant", "wolf", "fox", "horse",

    # Cities
    "paris", "london", "tokyo", "delhi", "mumbai", "beijing", "berlin", "rome",

    # Technology
    "computer", "laptop", "keyboard", "mouse", "internet", "software",
    "hardware", "smartphone",

    # Fruits
    "apple", "banana", "mango", "orange", "grape", "pineapple",

    # Vehicles
    "car", "bus", "train", "airplane", "bicycle", "truck"
]
```

```python
vectors = np.array([model[word] for word in words])
```

```python
tsne = TSNE(n_components=2, random_state=42, perplexity=5)
reduced_vectors = tsne.fit_transform(vectors)
```

```python
plt.figure(figsize=(12, 8))

x = reduced_vectors[:, 0]
y = reduced_vectors[:, 1]

plt.scatter(x, y)

for i, word in enumerate(words):
    plt.annotate(word, (x[i], y[i]))

plt.title("t-SNE Visualization of Word Embeddings")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.grid(True)
plt.show()
```

The t-SNE visualization shows clear clustering among semantically similar words.
Animal-related words such as cat, dog, lion, and tiger appear close together, indicating that their embeddings share similar contextual meanings.
City names like Paris, London, and Tokyo also form a visible cluster.
Technology-related words such as computer, laptop, and keyboard are positioned near one another, forming another group.
Fruit names such as apple, banana, and mango appear grouped together, separate from vehicles and cities.
Vehicle-related words like car, bus, and train also cluster in their own region.
Some words may appear slightly outside their expected clusters due to overlapping contexts in real-world language usage.
For example, “apple” might appear slightly closer to technology words because it is also a technology brand.
Overall, the visualization confirms that word embeddings successfully capture semantic similarity.
